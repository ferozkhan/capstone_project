 ### Segmentation and Clustering Neighborhood in Toronto

```python
url_of_list_of_postal_code = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
columns = ["PostalCode", "Borough", "Neighborhood"]
```

```python
neighbor_data = pd.read_html(url_of_list_of_postal_code, attrs={"class": "wikitable"})[0]
neighbor_data.columns = columns
```

```python
filter_borough_not_assigned = neighbor_data['Borough'] != 'Not assigned'
```

```python
neighbor_data = neighbor_data[filter_borough_not_assigned]
```

#### Preparing dataframe with combine neighborhood for each postalcode

```python
from collections import defaultdict
dataset = []

def process_postal_code(postal_code):
    """
    function to combine neighborhood for the provided 
    postal code.
    """
    for d in k[postal_code]:
        _borough, _neighbouhood = set(), set()
        for _, borough in d['Borough'].items():
            _borough.add(borough)
        for _, neighborhood in d['Neighborhood'].items():
            _neighbouhood.add(neighborhood) 

    dataset.append({
        'PostalCode': postal_code, 
        'Neighborhood': ', '.join(_neighbouhood), 
        'Borough': ''.join(_borough)
    })


def pre_process_neighbor_data(neighbor_data):
    """
    function to pre-process neighbor data to find similar
    neighborhood for the postalcode.
    """
    k = defaultdict(list)
    for n in neighbor_data['PostalCode']:
        k[n].append(neighbor_data[neighbor_data['PostalCode'] == n][['Borough', 'Neighborhood']].to_dict())

    for postal_code in k:
        process_postal_code(postal_code)

pre_process_neighbor_data(neighbor_data)
dataframe = pd.DataFrame(dataset)
```

#### replacing any 'Not assigned' Neighborhood value with the value of Borough. 

```python
neighborhoor_not_assigned = dataframe.Neighborhood == 'Not assigned'
dataframe.loc[neighborhoor_not_assigned, 'Neighborhood'] = dataframe.loc[neighborhoor_not_assigned, 'Borough']
```
